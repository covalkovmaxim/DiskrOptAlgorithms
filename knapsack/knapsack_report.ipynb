```python
import matplotlib.pyplot as plt
from pylab import *
```

## Timework competition

```python
fig, ax = plt.subplots(figsize=(15, 10))
with open('integer_linprog_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x,label="linprog")
print("Total worktime of integer linear programing: "+str(sum(x)))
with open('genetic_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of genetic algorithm: "+str(sum(x)))
with open('dinamic_2_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of dinamic by prices: "+str(sum(x)))
with open('dinamic_1_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of dinamic by wieghts: "+str(sum(x)))
with open('greedy_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of greedy algorithm: "+str(sum(x)))
with open('approximate_1_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of approximate algorithm, where eps=0.1: "+str(sum(x)))
with open('approximate_2_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of approximate algorithm, where eps=0.01: "+str(sum(x)))
ax.legend(["intlinprog","genetic","dinamic2","dinamic1","greedy","approx-0.1","approx-0.01"],loc='upper right')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
with open('greedy_accuracy.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
print("Mean accuracy="+str(mean(x)))
subplot(221)
title('Accuracy Histogram')
plt.hist(x,bins=100)
subplot(222)
title('Accuracy Graphic')
plt.plot(x)
plt.plot([mean(x)]*200)
with open('greedy_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
subplot(212)
title('Timework Graphic')
plt.plot(x)
plt.show()
print("Mean timework="+str(mean(x)))
```
