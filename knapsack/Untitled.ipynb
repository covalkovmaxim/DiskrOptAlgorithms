```python
import matplotlib.pyplot as plt
from pylab import *
```

## Timework competition

```python
fig, ax = plt.subplots(figsize=(15, 10))
with open('integer_linprog_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x,label="linprog")
print("Total worktime of integer linear programing: "+str(sum(x)))
with open('genetic_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of genetic algorithm: "+str(sum(x)))
with open('dinamic_2_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of dinamic by prices: "+str(sum(x)))
with open('dinamic_1_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of dinamic by wieghts: "+str(sum(x)))
with open('greedy_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of greedy algorithm: "+str(sum(x)))
with open('approximate_1_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of approximate algorithm, where eps=0.1: "+str(sum(x)))
with open('approximate_2_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
plt.plot(x)
print("Total worktime of approximate algorithm, where eps=0.01: "+str(sum(x)))
ax.legend(["intlinprog","genetic","dinamic2","dinamic1","greedy","approx-0.1","approx-0.01"],loc='upper right')

plt.show()
```

Как и ожидалось, жадный алгоритм оказался наиболее быстрым алгоритмом из участвовавших в сравнении, чем, в некотором роде, компенсировал свою неточность. <br>
Из точных алгоритмов победителем оказалась динамика по весам, а в аутсайдерах данного соревнования- целочисленное линейное программирование и генетический алгоритм. <br>
Перейдём к более детальному рассмотрению реализованных алгоритмов.

## Geedy Algorithm

Наверное, наиболее простой алгоритм из рассматриваемых. Теоретически дает 0.5-точность.

```python
fig, ax = plt.subplots(figsize=(15, 10))
with open('greedy_accuracy.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
print("Mean accuracy="+str(mean(x)))
subplot(221)
title('Accuracy Histogram')
plt.hist(x,bins=100)
subplot(222)
title('Accuracy Graphic')
plt.plot(x)
plt.plot([mean(x)]*200)
with open('greedy_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
subplot(212)
title('Timework Graphic')
plt.plot(x)
plt.show()
print("Mean timework="+str(mean(x)))
```

Как видим, на данных тестах жадный алгорим показал себя хорошо: его точность в среднем составляла 0.9  <br>
Реализация алгоритма находится в <b>knap.cpp</b> в функции <b>greedy_algorithm</b>

## Dinamic algorithm by prices

Точный алгоритм. Сложность $O(nW)$. Требуется столько же дополнительной памяти. <br>
Если со сложностью все не так страшно, то память здесь критична- даже для средних по размеру задач у меня не получается выделить столько памяти-badalloc.   

```python
fig, ax = plt.subplots(figsize=(15, 10))
with open('dinamic_1_accuracy.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
subplot(221)
print("Mean accuracy="+str(mean(x)))
title('Accuracy Histogram')
plt.hist(x,bins=100)
subplot(222)
title('Accuracy Graphic')
plt.plot(x)
plt.plot([mean(x)]*200)
with open('dinamic_1_time.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line) for line in lines]
subplot(212)
title('Timework Graphic')
plt.plot(x)
plt.show()
print("Mean timework="+str(mean(x)))
```

Показывает наилучшие результаты среди точных алгоритмов на данных тестах. <br>
Реализация алгоритма находится в <b>knap.cpp</b> в функции <b>dinamic_algorithm_1</b>

## Results

Ввиду неполноты тестов, однозначно назвать лучший алгоритм, конечно, нельзя. Так, аутсайдеры тестирования: целочисленное линейное программирование и генетический алгоритм не обладают существенными ограничениями по памяти и применимы для решения задач, о которых динамическим алгоритмам с их объемными матрицами динамики не приходиться даже мечтать. <br>
Однако одно можно сказать с полной увереностью-универсального алгоритма, работающего наилучшим образом во всех возможных ситуациях для задачи ркзака на данный момент - нет. Лучшим является комбинация алгоритмов, учитывающая различные особенности задачи.
